```python
#we know the 1mb = 1024kb

def size_process(item):
     if str(item)[-1] == 'm': # the value is m find
         res =float(str(item).replace("m", "")) #convert the value in to impacity space 
         res = res *1024
         return res
     elif str(item)[-1] == 'k': #if the value is k in last 
         res =float(str(item).replace("k", ""))#k also change the value to emporty sate
         return res
     else:
         return str(np.nan)# for any strung vlaue wreote in the data set convert into the null value
```

```python
#for cheak the data is correct or not 
size_process("19m")
#yes our data is work correctly
```
